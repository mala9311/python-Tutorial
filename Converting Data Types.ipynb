Converting Data Types

```python
num_int = 7
print(num_int)
```

```python
num_str = '7'
print(num_str)
```

```python
type(num_str)
```

```python
num_str_conv = int(num_str)
print(num_str_conv)
```

```python
type(num_str_conv)
```

```python
int_sum = num_int + num_str_conv
print(int_sum)
```
