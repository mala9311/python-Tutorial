For Loops

```python
integers = [1,2,3,4,5]
for jelly in integers:
    print(jelly +jelly)
```

```python
ice_cream_dict ={'Name': 'Mala Kumari', 'weekly intake': 5, 'Favorite ice cream': ['MCC' , 'Chocolate']}
for cream in ice_cream_dict.values():
    print(cream)
```

```python
for key, value in ice_cream_dict.items():
    print(key,'->',value)
```
