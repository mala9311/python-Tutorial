# Sets
-> sets provide unique value

```python
daily_pints = {1,2,3}

type(daily_pints)
```

```python
print(daily_pints)
```

```python
daily_pints_log = {1,2,3,4,3,2,1,5,32,1,2,3,4}
print(daily_pints_log)
```

```python
wifes_daily_pints_log = {1,2,4,5,67,8,5,3,21,2,1,19,0}
```

```python
# Provie all unique value
print(daily_pints_log | wifes_daily_pints_log  )
```

```python
# provide union 
print(daily_pints_log & wifes_daily_pints_log  )
```

```python
# provide what does not match with wifes_daily_pints_log 
print(daily_pints_log - wifes_daily_pints_log  )
```

```python
# provide what does not match with daily_pints_log 
print(wifes_daily_pints_log - daily_pints_log  )
```

```python
# provide intersection means the values from both the variable that all values that does not match

print(daily_pints_log ^ wifes_daily_pints_log  )
```
